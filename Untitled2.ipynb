```python
import pandas as pd
import numpy as np
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
df1 = pd.read_csv('/content/kecepatan.csv')
speed = df1[['speed']]
time = df1[['Time']]
X_train, X_test, y_train, y_test = ms.train_test_split(speed, time, test_size=0.2, random_state=0)
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('Speed')
plt.ylabel('Time')
plt.title('Kecepatan Kendaraan Perwaktu')
x1 = np.linspace(0,45)
y1 = 10.641 + 6.458 * x1
plt.plot(x1,y1)
plt.show()
```

```python
r2 = model1.score(X_test, y_test)
print(r2)
```
